```python
# import the package
import numpy as np
import matplotlib.pyplot as plt
import control as ct
from control.matlab import *
```

```python
# Example 1 - Cruise Control
#         ___________ 
#        |           |            x'
#    Bx' |           |    u     ----->
#  <-----|           |--------> 
#        |           |
#        |___________|
#        ooooooooooooo
#
m = 1000  # kg
B = 50   # N.s/m
u = 500  # N

nun = [1/m]
den = [1,B/m]

G = ct.tf(nun,den)
G
```

```python
# plot
y, t = step(u*G)
plt.plot(t,y)
plt.ylabel('Velocity $v$ [m/s]')
plt.xlabel('Time $t$ [s]')
```

```python
# Example 2 - Pendulum
t = np.linspace(0, 10, num=501)  # vector of times for output, 0 to 10 at 0.02
m = 1          # kg
L = 1          # m
g = 9.81       # m/s^2
angle = 5      # degree
s = tf('s')    # s (Laplace variable)

# TF
H = ( 1 / (m*L**2) ) / (s**2 + g/L)
y, T = step(H,t)
angle_rad = np.deg2rad(angle)

```

```python
plt.plot(T,angle_rad*y)
plt.ylabel('Pendulum Angle')
plt.xlabel('Time $t$ [s]')
```
